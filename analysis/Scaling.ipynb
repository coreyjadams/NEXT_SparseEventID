# 512 Images per GPU 

```python
x = np.array([1, 2, 4, 8])
print x
iotest    = np.array([1.7, 3.7, 6.8, 13.9])
# io        = np.array([1.7, 3.7, 6.8])
delta_log = np.array([6.1, 8.5, 11.5, 18.7])

# x_ticks_labels = ['1','2','4']

fig, ax = plt.subplots(1,1) 

ax.plot(x, iotest, marker="o", label='IO Time')
ax.plot(x, delta_log, marker="o", label='Delta Log Time')

ax.legend()
ax.set_xlabel(r'Number of GPUs', fontsize=15, labelpad=10)
ax.set_ylabel(r'Time', fontsize=15, labelpad=10)

ax.set_xticks(x)
# ax.set_xticklabels(x_ticks_labels, rotation='vertical', fontsize=18)
plt.savefig('io_deltalog_512pergpu.pdf')
```
